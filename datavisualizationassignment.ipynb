```python
import matplotlib.pyplot as plt
data=pd.read_csv(r"D:\apple.csv")

# Bar Chart: MRP vs Sale Price (Top 10 Products)
plt.figure(figsize=(10, 5))
plt.bar(data["Product Name"][:10], data["Mrp"][:10], label="MRP", color='gray', alpha=0.7)
plt.bar(data["Product Name"][:10], data["Sale Price"][:10], label="Sale Price", color='blue', alpha=0.7)

# Customizing the chart
plt.xticks(rotation=45, ha="right", fontsize=9)  
plt.title("MRP vs Sale Price (Top 10 Products)", fontsize=14)
plt.xlabel("Product Name")
plt.ylabel("Price (INR)")
plt.legend()
plt.tight_layout()  # Adjust layout to fit labels properly

# Display the plot
plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
data=pd.read_csv(r"D:\apple.csv")

# Select the first 5 products and their sale prices
products = data["Product Name"][:5]
prices = data["Sale Price"][:5]

# Create a bar chart
plt.bar(products, prices, color='blue', alpha=0.8)

# Add labels and title
plt.xlabel("Products")  # Label for x-axis
plt.ylabel("Sale Price (INR)")  # Label for y-axis
plt.title("Sale Prices of Top 5 Products")  # Title of the chart

# Display the chart
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to fit labels
plt.show()
```

```python
plt.figure(figsize=(7, 7))
plt.pie(prices, labels=products, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'orange', 'lightgreen', 'pink', 'yellow'])
plt.title("Sale Price Distribution (Top 5 Products)")
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Extract Sale Price data
sale_prices = data["Sale Price"]

# Create a histogram of Sale Prices
plt.figure(figsize=(8, 6))
plt.hist(sale_prices, bins=10, color='skyblue', edgecolor='black', alpha=0.7)

# Add labels and title
plt.xlabel("Sale Price (INR)")
plt.ylabel("Frequency")
plt.title("Distribution of Sale Prices")

# Display the histogram
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Extract Product Name and Sale Price data
products = data["Product Name"]
sale_prices = data["Sale Price"]

# Create a line plot
plt.figure(figsize=(10, 5))
plt.plot(products, sale_prices, marker='o', color='blue', linestyle='-', alpha=0.7)

# Add labels and title
plt.xlabel("Product Name")
plt.ylabel("Sale Price (INR)")
plt.title("Sale Prices of Products")

# Rotate x-axis labels for readability
plt.xticks(rotation=45, ha="right")

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file

# Extract Number of Ratings and Sale Price data
ratings = data["Number Of Ratings"]
sale_prices = data["Sale Price"]

# Create a scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(ratings, sale_prices, color='green', alpha=0.7)

# Add labels and title
plt.xlabel("Number of Ratings")
plt.ylabel("Sale Price (INR)")
plt.title("Scatter Plot of Ratings vs Sale Price")

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Extract Sale Price data
sale_prices = data["Sale Price"]

# Create a box plot for Sale Prices
plt.figure(figsize=(8, 5))
plt.boxplot(sale_prices, vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue', color='black'))

# Add labels and title
plt.xlabel("Sale Price (INR)")
plt.title("Box Plot of Sale Prices")

# Display the plot
plt.tight_layout()
plt.show()
```
